```python
#use this cell to begin your analyses, and add as many cells

import pandas as pd


theme = pd.read_csv('parent_themes.csv')

theme.head(5)




```

```python
merged_df = df.merge(theme, left_on='parent_theme', right_on='name')
merged_df.head()
```

```python
merged_df[merged_df['set_num'].isnull()].shape
```

```python
licensed = merged_df[merged_df['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])
licensed.head()

star_wars = licensed[licensed['parent_theme']=='Star Wars']

star_wars.head(5)

the_force = int(star_wars.shape[0]/licensed.shape[0]*100)

print(the_force)
print(f"{the_force}%")
```

```python
df_sorted = licensed.sort_values('year')
df_sorted.head()
```

```python
df_sorted = licensed.sort_values('year')
df_sorted['count'] = 1
summed_df = df_sorted.groupby(['year','parent_theme']).sum().reset_index()
summed_df.sort_values('count', ascending=False).drop_duplicates(['year'])
```
